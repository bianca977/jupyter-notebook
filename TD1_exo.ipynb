```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
temp=pd.read_csv("temperature-quotidienne-regionale.csv",sep=";")
```

```python
temp
```

```python
temp.columns=["Date","Insee","Region","TMin","TMax","TMoy"]
```

```python
temp
```

```python
#departement ou la temperature max a depasse les 40 degres
temp[temp["TMax"]>40]
```

```python
#departements ou la temperature min etait moins que -5 degres
temp[temp["TMin"]<-5]
```

```python
#tri du tableau par nom de region
temp.sort_values(by="Region")
```

```python
#20 departements ou la temp moy quotidienne etait le plus forte
temp.sort_values(by="TMoy")[-20:]
```

```python
#les 10 temp max enregistrees en corse 
temp[temp.Region="Corse"].sort_values(by='TMax')[-10:]
```

```python
#les 10 temp max enregistrees en corse 
temp[temp.Region=="Corse"].sort_values(by='TMax')[-10:]
```

```python
#top 10 des temp min enregistrees en Ile de France
temp[temp.Region=="Ile-de-France"].sort_values(by='TMin')[-10:]
```

```python
#top des temp min enregistrees en Ile de France
temp[temp.Region=="Ile-de-France"].sort_values(by='TMin',ascending=False)
```

```python
#Top des departements avec 
temp.groupby("Region").TMoy.mean().sort_values[-20:]
```

```python
temp.groupby("Region").TMoy.sum().sort_values[-20:]
```
